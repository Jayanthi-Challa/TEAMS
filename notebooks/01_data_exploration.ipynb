```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
population = pd.read_csv("../data/raw/tiger_population.csv")
funding = pd.read_csv("../data/raw/reserve_funding.csv")
poaching = pd.read_csv("../data/raw/poaching_incidents.csv")
patrol = pd.read_csv("../data/raw/patrol_paths.csv")
```

```python
population.head(), funding.head(), poaching.head(), patrol.head()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=population, x="year", y="population", hue="reserve_id")
plt.title("Tiger Population Trend Over Time")
plt.show()
```

```python
merged = population.merge(funding, on="reserve_id", how="left")
merged.head()
```

```python
merged.to_csv("../data/processed/merged_data.csv", index=False)
```
